```python
import pandas as pd
```

```python
full_data = pd.read_csv('fulldata.csv')
train_data = pd.read_csv('traindata.csv')
test_data = pd.read_csv('testdata.csv')
```

```python
full_len = len(full_data)
print('full_len:', full_len)
train_len = len(train_data)
print('train_len:', train_len)
test_len = len(test_data)
print('test_len:', test_len)
```

```python
list = []
for ti,trow in test_data.iterrows():
    print(ti)
    found=False
    for fi,frow in full_data.iterrows():
        if frow.drop('A16').equals(trow):
            print('found', frow.tolist())
            found=True
            full_data.drop(index=fi, axis=0, inplace=True)
            #test_data.iloc[ti]=frow
            #testfullset=pd.concat([testfullset,frow], ignore_index=True)
            list.append(frow['A16'])
            break
    if not found:
        print('not found')
        list.append('Not Found')
```

```python
test_y = pd.Series(list)
test_data['A16'] = test_y
test_data.tail()
```

```python
test_data.to_csv('full_test.csv')
```

```python
test_data = pd.concat([full_data, train_data]).drop_duplicates(keep=False)
print(len(test_data))
```

```python
test_data2 = full_data.merge(train_data, how = 'outer' ,indicator=True).loc[lambda x : x['_merge']=='left_only']
test_data2.to_csv('test_full_new.csv')
print(len(test_data2))
```
